```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
with open("./data/wiki_us.txt") as f :
    text = f.read()
```

```python
print(text)
```

```python
doc = nlp(text)
```

```python
print(doc)
```

```python
# simple iterible text of the document file but now we cover difference between original text and doc container first we
# start iterible text
for text in text[:10]:
    print(text)
```

```python
# token.text mean access raw text of token metadata
for token in doc[:10]:
    print(token.text)
```

```python
# now we find part of speech first of 10 tokens
for token in doc[:10]:
    print(f'{token} : {token.pos_}')
```

```python
# now we find syntactic structure sentence inlcuding with their relationship tokens
for token in doc[:10]:
    print(f'{token} : {token.dep_}')
```

```python
# lemmitazation first 20 index of doc container in a each token or reducing of each token
for token in doc[:20]:
    print(f'{token} : {token.lemma_}')
```

```python
# language of each token in a first 10 index token of doc contianer
for token in doc[:10]:
    print(f'{token} : {token.lang_}')
```

```python
# token entity types of first index of doc contianer
for token in doc[:10]:
    print(f'{token} : {token.ent_type_}')
```

```python
# token inside , outside , beginner entity type of first index of doc contianer
for token in doc[:10]:
    print(f'{token} : {token.ent_iob_}')
```

```python
for token in doc[:10]:
    print(token.left_edge)
```

```python
for token in doc[:10]:
    print(token.right_edge)
```
